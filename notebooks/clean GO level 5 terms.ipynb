```python
import pandas as pd
import numpy as np
```

```python
genes = pd.read_csv("/Users/veronica/Dropbox (DIBRIS)/project_UCL/level5GOterms.csv")
go_terms = pd.read_csv("/Users/veronica/Dropbox (DIBRIS)/project_UCL/GOlevel5.csv")
go_terms = list(go_terms['x'])
```

```python
df = pd.DataFrame([[np.nan]], columns=['x'], index=['x'])
genes = df.append(genes)
```

```python
go_term = 0
df_res = pd.DataFrame(columns=['Gene', 'GOterm'])
i=0
while i < genes.shape[0]:
    try:
        go = go_terms[go_term]
    except:
        break
    go_term +=1
    for j in range(i+1, genes.shape[0]):
        if genes.index.values[j] == 'x':
            i = j
            break
        df = pd.DataFrame([[genes.iloc[j, 0], go]], columns=['Gene', 'GOterm'])
        df_res = df_res.append(df, ignore_index=True)
```

```python
df_res = df_res.set_index('Gene')
df_res.index = [str(g).lower() for g in df_res.index]
df_res
```

```python
df_res.to_csv("/Users/veronica/Dropbox (DIBRIS)/project_UCL/db_GO_level5_big.csv")
```
